При запуске кода каждый раз получаем новый набор из трёх фотографий:

```python
import requests
import matplotlib.pyplot as plt 
import numpy as np
from skimage import io

key = 'jPP1qlpe4oiHoW2MWdGvbDHU4XKzOWhgGKBIkJOj'
url = f'https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?earth_date=2022-2-6&api_key={key}'

response = requests.get(url) 
data = response.json()
arr_with_index = np.random.randint(0, 456, size=3)

fig, ax = plt.subplots(figsize=(10, 20), nrows=3, ncols=1)
for i in range(len(arr_with_index)):
    url_img = data['photos'][arr_with_index[i]]['img_src']
    img = io.imread(url_img)
    ax[i].imshow(img, cmap='gray')
    
for axi in ax.ravel(): 
    axi.set_axis_off()
```
